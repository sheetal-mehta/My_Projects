```python
import pandas as pd
import math
import random
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('lr.csv')
```

```python
df = pd.DataFrame(data)
```

```python
shapee = df.shape
```

```python
row = shapee[0]
columnn = shapee[1] 
```

```python
def mse_calc(m,c):
    mse = 0
    for i in range(row):
        mse = mse+(((df.iloc[i,2]-(m*df.iloc[i,1])+c))**2)
    mse = mse/row
    return mse
```

```python
m = random.randrange(1,100)
c = random.randrange(1,100)
mse_calc(m,c)
```

```python
def mm(m,c):
    mm = 0
    for i in range(row):
        mm = mm+((m*df.iloc[i,1])+c-df.iloc[i,2])*df.iloc[i,1]
    mm = (2*mm)/row
    return mm  
```

```python
def cc(m,c):
    cc = 0
    for i in range(row):
        cc = cc + ((m*df.iloc[i,1])+c-df.iloc[i,2])
    cc = (2*cc)/row
    return cc
```

```python
alpha_1 = 0.01
alpha_2 = 0.0001
```

```python
mse_list = []
for i in range(100000):
    if i <10000:
        m = m-(alpha_1*mm(m,c))
        c = c-(alpha_1*cc(m,c))
    else:
        m = m-(alpha_2*mm(m,c))
        c = c-(alpha_2*cc(m,c))

    mse_list.append(mse_calc(m,c))
print(mse_list)
```

```python
l = [i for i in range(100000)]
plt.plot(l,mse_list)
```
