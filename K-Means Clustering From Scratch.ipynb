```python
# getting the number of clusters by user
n = int(input('Enter the value of clusters: '))
```

```python
# Obtaining the size of dataframe to iterate through rows and columns
df_size = df.shape
```

```python
df_column_names = df.columns
```

```python
column_size = df_size[1]
row_size = df_size[0]
```

```python
# variables to store centroids
clusters = []
for i in range(0,n):
    clusters.append('C'+str(i))
clusters
```

```python
# Lower limit to generate centroid point
min_range = {df_column_names[i]:min(df[df_column_names[i]]) for i in range(len(df_column_names))}
min_range
```

```python
# upper limit to generate centroid point
max_range = {df_column_names[i]:max(df[df_column_names[i]]) for i in range(len(df_column_names))}
max_range
```

```python
# creating the centroid
centroids = {clusters[i]:[random.uniform(min_range[df_column_names[j]],max_range[df_column_names[j]])for j in range(0,column_size)] for i in range(0,n)} 
```

```python
# this function takes dataframe, centroid and number of clusters and returns the clusters
def cluster(df, C,n):
    clust = []
    clusters = []
    for i in range(0,n):
        clusters.append('C'+str(i))
    dist = ['d'+str(i) for i in range(n)]
    
    # this loops calculates the distances against all the centroids and saves them in a list.
    for i in range(len(C)):
        dist[i] = []
        for j in range(row_size):
            summ = 0
            for k in range(column_size):
                summ = summ + ((df.iloc[j,k]-C[clusters[i]][k])**2)
            dist[i].append(math.sqrt(summ))

    # This loop is for the comparision of the distances and accordingly assigns the cluster
    print(dist[0])
    print(dist[1])
    for i in range(row_size):
        bigg = dist[0][i]
        print('i:',i)
        for j in range(n):
            print('j:',j)
            if dist[j][i]<=bigg:
                bigg = dist[j][i]
                index = j
                print(index)
        clust.append(index)
    return clust
```

```python
df['cluster'] = cluster(df,centroids,n)
```
