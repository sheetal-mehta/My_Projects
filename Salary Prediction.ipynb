```python
def prob_less_than_50K(row):
    return (1/(math.sqrt(2*math.pi*(X_con_sdv_less['sdv_age_less'])**2))*math.exp(-((row['age']-X_con_mean_less['mean_age_less'])**2)/(2*(X_con_sdv_less['sdv_age_less']**2))))*(1/(math.sqrt(2*math.pi*(X_con_sdv_less['sdv_ fnlwgt_less'])**2))*math.exp(-((row[' fnlwgt']-X_con_mean_less['mean_ fnlwgt_less'])**2)/(2*(X_con_sdv_less['sdv_ fnlwgt_less']**2))))*(1/(math.sqrt(2*math.pi*(X_con_sdv_less['sdv_ education-num_less'])**2))*math.exp(-((row[' education-num']-X_con_mean_less['mean_ education-num_less'])**2)/(2*(X_con_sdv_less['sdv_ education-num_less']**2))))*(1/(math.sqrt(2*math.pi*(X_con_sdv_less['sdv_ capital-gain_less'])**2))*math.exp(-((row[' capital-gain']-X_con_mean_less['mean_ capital-gain_less'])**2)/(2*(X_con_sdv_less['sdv_ capital-gain_less']**2))))*(1/(math.sqrt(2*math.pi*(X_con_sdv_less['sdv_ capital-loss_less'])**2))*math.exp(-((row[' capital-loss']-X_con_mean_less['mean_ capital-loss_less'])**2)/(2*(X_con_sdv_less['sdv_ capital-loss_less']**2))))*(1/(math.sqrt(2*math.pi*(X_con_sdv_less['sdv_ hours-per-week_less'])**2))*math.exp(-((row[' hours-per-week']-X_con_mean_less['mean_ hours-per-week_less'])**2)/(2*(X_con_sdv_less['sdv_ hours-per-week_less']**2))))
```

```python
def prob_more_than_50K(row):
    return (1/(math.sqrt(2*math.pi*(X_con_sdv_more['sdv_age_more'])**2))*math.exp(-((row['age']-X_con_mean_more['mean_age_more'])**2)/(2*(X_con_sdv_more['sdv_age_more']**2))))*(1/(math.sqrt(2*math.pi*(X_con_sdv_more['sdv_ fnlwgt_more'])**2))*math.exp(-((row[' fnlwgt']-X_con_mean_more['mean_ fnlwgt_more'])**2)/(2*(X_con_sdv_more['sdv_ fnlwgt_more']**2))))*(1/(math.sqrt(2*math.pi*(X_con_sdv_more['sdv_ education-num_more'])**2))*math.exp(-((row[' education-num']-X_con_mean_more['mean_ education-num_more'])**2)/(2*(X_con_sdv_more['sdv_ education-num_more']**2))))*(1/(math.sqrt(2*math.pi*(X_con_sdv_more['sdv_ capital-gain_more'])**2))*math.exp(-((row[' capital-gain']-X_con_mean_more['mean_ capital-gain_more'])**2)/(2*(X_con_sdv_more['sdv_ capital-gain_more']**2))))*(1/(math.sqrt(2*math.pi*(X_con_sdv_more['sdv_ capital-loss_more'])**2))*math.exp(-((row[' capital-loss']-X_con_mean_more['mean_ capital-loss_more'])**2)/(2*(X_con_sdv_more['sdv_ capital-loss_more']**2))))*(1/(math.sqrt(2*math.pi*(X_con_sdv_more['sdv_ hours-per-week_more'])**2))*math.exp(-((row[' hours-per-week']-X_con_mean_more['mean_ hours-per-week_more'])**2)/(2*(X_con_sdv_more['sdv_ hours-per-week_more']**2))))
```

```python
X_test['num_prob_less'] = X_test.apply(prob_less_than_50K,axis = 1)
X_test
```

```python
X_test['num_prob_more'] = X_test.apply(prob_more_than_50K,axis = 1)
X_test
```

```python
def prob_cat_less(row):
    return p_cat_less[row[' workclass']]*p_cat_less[row[' education']]*p_cat_less[row[' marital-status']]*p_cat_less[row[' occupation']]*p_cat_less[row[' relationship']]*p_cat_less[row[' race']]*p_cat_less[row[' sex']]*p_cat_less[row[' native-country']]*target_probabilities[' <=50K']
```

```python
def prob_cat_more(row):
    try:
        return p_cat_more[row[' workclass']]*p_cat_more[row[' education']]*p_cat_more[row[' marital-status']]*p_cat_more[row[' occupation']]*p_cat_more[row[' relationship']]*p_cat_more[row[' race']]*p_cat_more[row[' sex']]*p_cat_more[row[' native-country']]*target_probabilities[' >50K']
    except:
        return 0
```

```python
X_test['cat_prob_less'] = X_test.apply(prob_cat_less,axis = 1)
```

```python
X_test['cat_prob_more'] = X_test.apply(prob_cat_more,axis = 1)
```

```python
X_test['Final prob_less'] = X_test['num_prob_less']*X_test['cat_prob_less']
X_test['Final prob_more'] = X_test['num_prob_more']*X_test['cat_prob_more']
# Verifiy if the multiplication is at row level and probabilities are coming out correct
```

```python
X_test
```

```python
def final_prediction(row):
    if row['Final prob_less']>row['Final prob_more']:
        return ' <=50K'
    else:
        return ' >50K'
```

```python
X_test['Predicted Salary'] = X_test.apply(final_prediction,axis = 1)
```

```python
X_test
```

```python
def check(row):
    if row[' Salary']==row['Predicted Salary']:
        return 1
    else:
        return 0
```
